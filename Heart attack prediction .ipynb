# Feature engineering 

```python
df.drop('education', axis=1 , inplace=True)
df.drop('diaBP', axis=1 , inplace=True)
```

```python
from sklearn.preprocessing import StandardScaler,Normalizer,LabelEncoder
from category_encoders.target_encoder import TargetEncoder
from sklearn.model_selection import train_test_split,cross_val_score,RandomizedSearchCV,StratifiedKFold
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier,ExtraTreeClassifier
from sklearn.ensemble import AdaBoostClassifier,GradientBoostingClassifier,RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
X = df.drop(['TenYearCHD'],axis=1)
y = df['TenYearCHD']
```

```python
scaler = Normalizer()
X = scaler.fit_transform(X)
```

```python
X
```

# Model building 

```python

 X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42,stratify=y)
```

```python

lr_model = LogisticRegression()
```

```python

lr_model.fit(X_train,y_train)
```

```python
lr_preds = lr_model.predict(X_test)
confusion_matrix(lr_preds,y_test)
```

```python
y_pred_train=lr_model.predict(X_train)
```

```python
y_pred_test=lr_model.predict(X_test)
```

```python

accuracy_score(y_train, y_pred_train)
```

```python
accuracy_score(y_test,y_pred_test)
```

```python
from sklearn.metrics import roc_curve,roc_auc_score
fpr, tpr, thresholds = roc_curve(y, lr_model.predict_proba (X)[:,1])

auc = roc_auc_score(y_train,y_pred_train)
print(auc)

import matplotlib.pyplot as plt
plt.plot(fpr, tpr, color='blue', label='logit model ( area  = %0.2f)'%auc)
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate or [1 - True Negative Rate]')
plt.ylabel('True Positive Rate')
```

```python
print(classification_report(lr_preds,y_test))
```

```python

rf_model = RandomForestClassifier()
```

```python
rf_model.fit(X_train,y_train)
```

```python
preds = rf_model.predict(X_test)
confusion_matrix(preds,y_test)
```

```python
y_pred_train=rf_model.predict(X_train)
```

```python
y_pred_test=rf_model.predict(X_test)
```
